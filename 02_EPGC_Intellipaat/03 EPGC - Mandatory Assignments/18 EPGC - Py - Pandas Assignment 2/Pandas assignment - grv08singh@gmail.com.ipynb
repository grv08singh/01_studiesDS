```python
A = pd.DataFrame([['Carl', 22],['Martha', 25],['Calvin', 12],['Stuart', 15]], columns=['Name', 'Age']) 
B = pd.DataFrame([['Melvin', 25],['Martha', 34],['Lewis', 32],['Leo', 25]], columns=['Name', 'Age']) 
```

```python
B
```

```python
#1. Left Outer Join 
pd.merge(A,B,how='left',on='Name')
```

```python
#2. Outer Join 
pd.merge(A,B,how='outer',on='Name')
```

```python
#3. Inner Join 
pd.merge(A,B,how='inner',on='Name')
```

```python
#4. Right Outer Join
pd.merge(A,B,how='right',on='Name')
```
