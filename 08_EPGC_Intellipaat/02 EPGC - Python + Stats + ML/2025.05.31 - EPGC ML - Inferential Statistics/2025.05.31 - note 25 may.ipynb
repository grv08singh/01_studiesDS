```python
#The average salary of a person is hundred dollars back when the survey was conducted 
#in 2020 in fresh survey conducted in 2025 with 30 participants, it was found out that 
#the average salary of a person is $87 with a standard deviation of $20 can you 
#calculate whether this is a significant difference in salary or not at a 
#confidence interval of 95%?
```

```python
#null(h0) : (mu = 100) (no significant change in salaries)
#alt(h1)  : (mu != 100) (significant changes in salaries)
```

```python
#lib
import numpy as np
from scipy import stats

#data
population_mean = 100 #2019
sample_mean = 87 #2025
n = 30
std_dev = 20
alpha = 0.05

#t_statistic
t_statistic = (sample_mean - population_mean)/(std_dev/np.sqrt(n))
print(f"t_statistic : {t_statistic}")

#degree of freedom
df = n - 1
print(f"df : {df }")

# #t_critical_Value percent point function
# #if 2 tail test - alpha/2.    if one tail
# t_critical_value_upperbound = stats.t.ppf(1 - alpha/2, df)
# t_critical_value_lowerbound = stats.t.ppf(alpha/2, df)
# print(f"t_critical_value_upperbound : {t_critical_value_upperbound}")
# print(f"t_critical_value_lowerbound : {t_critical_value_lowerbound}")


#if you impo 
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic),df)) #syntax
print(f"p_value : {p_value}")


if p_value < alpha:
    print("Reject H0 : (mu != 100) (significant changes in salaries)")
else:
    print("Accept H0: (mu = 100) (no significant change in salaries)")
```

```python
#one tail
karnataka_average = 70

sanatan_vishyalay_Score = [72,75,68,70,76,80,79,73,71,78] #real life


t_statistic, p_value = stats.ttest_1samp(sanatan_vishyalay_Score, karnataka_average)
print(f"t_statistic: {t_statistic}")

#one tail   #two tail
p_value = p_value / 2
print(f"p_value : {p_value}")

alpha= 0.05

if p_value < alpha:
    print("Reject H0 :  s school > karn")
else:
    print("Accept H0: kar = s school")
```
